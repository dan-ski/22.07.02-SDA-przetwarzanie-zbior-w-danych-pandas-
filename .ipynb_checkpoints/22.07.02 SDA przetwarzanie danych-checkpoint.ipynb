# Creating and deleting columns

```python
movie_path = 'https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/movie.csv'
movie = pd.read_csv(movie_path)
movie.head()
```

```python
#movie.drop(columns = ['color'])
movie = movie.drop(columns = ['color']) # == movie.drop(['director_name'], axis = 1)
movie.head()
```

```python
movie.drop(['director_name'], axis = 1)
movie.drop(index = [1,2,3]) # usunięcie wierszy 
```

```python
movie = movie.set_index('movie_title')
movie.drop(index = ['Avatar'])
```
