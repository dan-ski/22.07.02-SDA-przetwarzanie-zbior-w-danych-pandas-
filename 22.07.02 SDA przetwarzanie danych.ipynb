## Boolean Indexing

```python
movie_path = 'https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/movie.csv'
movie = pd.read_csv(movie_path)
movie.head()
```

```python
movie_2_h = movie['duration'] > 120
movie[movie_2_h]

# movie[movie['duration']> 120]
```

```python
movie_2_h = movie['duration'] > 120
movie_1_h = movie['duration'] < 60

movie_filter = movie_2_h | movie_1_h #  or

movie[movie_filter]
```

```python
movie_2_h = movie['duration'] > 120
movie_color = movie['color'] == 'Color'

movie_filter = movie_2_h & movie_color # and
movie_filter
```

```python
movie[movie_2_h & movie_color]
```

# Merge, Join, Concat

```python
from IPython.display import display

df1 = pd.DataFrame(
        {
            "A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],
            "C": ["C0", "C1", "C2", "C3"],
            "D": ["D0", "D1", "D2", "D3"],
        },
        index=[0, 1, 2, 3],
    )
    

df2 = pd.DataFrame(
        {
            "A": ["A4", "A5", "A6", "A7"],
            "B": ["B4", "B5", "B6", "B7"],
            "C": ["C4", "C5", "C6", "C7"],
            "D": ["D4", "D5", "D6", "D7"],
        },
        index=[4, 5, 6, 7],
    )
    

df3 = pd.DataFrame(
        {
            "A": ["A8", "A9", "A10", "A11"],
            "B": ["B8", "B9", "B10", "B11"],
            "C": ["C8", "C9", "C10", "C11"],
            "D": ["D8", "D9", "D10", "D11"],
        },
        index=[8, 9, 10, 11],
    )
display(df1)
display(df2)
display(df3)
```

```python
results = pd.concat([df1, df2, df3], keys = ['x', 'y', 'z']) # multiindex
results
```

```python
results.loc['y']
```

```python
df4 = pd.DataFrame(
        {
            "B": ["B2", "B3", "B6", "B7"],
            "D": ["D2", "D3", "D6", "D7"],
            "F": ["F2", "F3", "F6", "F7"],
        },
        index=[2, 3, 6, 7],
    )
```

```python
pd.concat([df1, df4])
```

```python
pd.concat([df1, df4], axis = 1)
```
